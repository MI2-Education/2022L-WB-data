# Chest X-ray images Open-i

```python
# ramka stworzona przy EDA

df = pd.read_csv('df_text.csv').drop(['Unnamed: 0', 'COMPARISON'], axis = 1)

df
```

```python
# szukanie płci w tekstach, balans płci

df['INDICATION'] = df['INDICATION'].fillna('')
df['FINDINGS'] = df['FINDINGS'].fillna('')
df['IMPRESSION'] = df['IMPRESSION'].fillna('')


df['sex_ind'] = df['INDICATION'].apply(
    lambda x : np.where(bool(re.search('female', x)), -1, np.where(bool(re.search(r' male|^male', x)), 1, 0))
)

df['sex_fin'] = df['FINDINGS'].apply(
    lambda x : np.where(bool(re.search('female', x)), -1, np.where(bool(re.search(r' male|^male', x)), 1, 0))
)

df['sex_imp'] = df['IMPRESSION'].apply(
    lambda x : np.where(bool(re.search('female', x)), -1, np.where(bool(re.search(r' male|^male', x)), 1, 0))
)

df['sex'] = df['sex_ind'] + df['sex_fin'] + df['sex_imp']

df['sex'] = df['sex'].replace({
    -1 : 'female',
    0  : 'no data',
    1  : 'male'
})
```

```python
df['sex'].value_counts()
```

```python
ax = sns.barplot(data = pd.DataFrame(df['sex'].value_counts()).T.reset_index(drop = True), color = 'dodgerblue')
ax.set(xlabel = 'płeć', ylabel = 'liczba', title = 'Balans płci')
plt.show()
```

```python
# analiza pozycji zdjęć RTG

df['CAPTION'] = df['CAPTION'].fillna('')
df['CAPTION'] = df['CAPTION'].apply(lambda x : x.lower())

df['position'] = df['CAPTION'].apply(
    lambda x : np.where(bool(re.search(r'lateral|pa', x)), 'PA', 
                                np.where(bool(re.search(r'2 view|2 chest|2|two', x)), '2 views', 
                                        np.where(bool(re.search(r'ap|portable', x)), 'AP', 'no data')))
)
```

```python
df['position'].value_counts()
```

```python
ax = sns.barplot(data = pd.DataFrame(df['position'].value_counts()).T.reset_index(drop = True), color = 'dodgerblue')
ax.set(xlabel = 'pozycja', ylabel = 'liczba zdjęć', title = 'Pozycje wykonywania zdjęć')
plt.show()
```

```python
from nltk.sentiment import SentimentIntensityAnalyzer
sia = SentimentIntensityAnalyzer()
```

```python
def sent(col):
    sentiment_ind = pd.DataFrame(columns = ['neg', 'neu', 'pos', 'compound'])

    for i in df[col]:
        x = sia.polarity_scores(i)
        sentiment_ind = pd.concat([sentiment_ind, pd.DataFrame(x, index = [1])])


    fig, axs = plt.subplots(2, 2, figsize=(15, 10))
    sns.histplot(sentiment_ind['neg'], ax = axs[0, 0])
    sns.histplot(sentiment_ind['neu'], ax = axs[0, 1])
    sns.histplot(sentiment_ind['pos'], ax = axs[1, 0] )
    sns.histplot(sentiment_ind['compound'], ax = axs[1, 1])
    fig.suptitle(col)
    plt.show()
```

### Wiek pacjentów

```python
def yrs_to_int(yrs):
    if type(yrs) != type([]): return np.nan
    if len(yrs) == 0: return np.nan
    return yrs[0].replace('yrs', '')

ages = df_china['content'].str.findall(r"[0-9]*yrs").apply(yrs_to_int).dropna()
ages = ages[ages != ""].astype('int')
sns.boxplot(data = ages)
```

### Wiek mężczyzn

```python
ages = df_china['content'][df_china['content'].str.contains('male')].str.findall(r"[0-9]*yrs").apply(yrs_to_int).dropna()
ages = ages[ages != ""].astype('int')
sns.boxplot(data = ages)
```

### Wiek kobiet

```python
ages = df_china['content'][df_china['content'].str.contains('female')].str.findall(r"[0-9]*yrs").apply(yrs_to_int).dropna()
ages = ages[ages != ""].astype('int')
sns.boxplot(data = ages.values)
```

Rozkład wieku podobny dla mężczyzn i kobiet.

```python
from collections import Counter
pd.DataFrame([
    list(elem) for elem in Counter(" ".join(df_china['content'].str.split("\n", n = 1).apply(lambda x: x[1])).split()).most_common(1000)
], columns = ['word', 'count']).head(25)
```
